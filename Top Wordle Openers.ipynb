### Linear Regression

A better approach is to control for the presence of other words, and look for an independent signal. If BRAZE only does well when it's paired with GRACE or GRADE or SHARE than a linear regression should isolate the guesses that actually are predictive of a popular score count line.

The difficulty comes from getting the dataframe into the right format. There must be an easier way to do this, but I need one row per score per wordle number with columns that are all the possible guesses with a one hot encoded data for each column: 1 if the word is in that score, and 0 if it is not. I did this via a groupby, sum, and loop.

If there is a better way to encode this categorical data in this grouped way, please let me know! (I think the answer is changing the initial dataframe so one column contains a list of the guesses, then using a `DictVectorizer` or something like it)

```python
from tqdm.notebook import tqdm
df.rename(columns={'score':'score_pattern'},inplace=True)
dummies = pd.get_dummies(df['guess']) # get the 

temp_df = pd.concat(
    [df[['wordle_num', 'score_count_fraction', 'guess_count','score_pattern']], dummies],
    axis=1)

summed_dfs = []
for grouping_vars, _df in tqdm(temp_df.groupby(['wordle_num','score_pattern'])):
    wordle_num, score_pattern = grouping_vars
    temp = _df[dummies.columns].sum(axis=0)
    temp['wordle_num'] = wordle_num
    temp['score_count_fraction'] = _df['score_count_fraction'].iloc[0]
    temp['guess_count'] = _df['guess_count'].iloc[0]
    summed_dfs.append(temp)
one_hot_encoded_data = pd.DataFrame(summed_dfs)
```

```python
one_hot_encoded_data
```

Now we'll normalize the data a little and then fit a model. We use a [Ridge](https://scikit-learn.org/stable/modules/linear_model.html#ridge-regression-and-classification) which helps control the size of the coefficients.

```python
one_hot_encoded_data['score_count_fraction'] = one_hot_encoded_data[
    'score_count_fraction'].fillna(0)
one_hot_encoded_data['score_count_fraction'] = one_hot_encoded_data[
    'score_count_fraction'] / one_hot_encoded_data[
        'score_count_fraction'].mean()
one_hot_encoded_data['guess_count'] = one_hot_encoded_data[
    'guess_count'] / one_hot_encoded_data['guess_count'].mean()
```

```python
from sklearn import linear_model
data = one_hot_encoded_data 
X= data.drop(
        columns=[ 'score_count_fraction', 'wordle_num']) # fit to the one hot encouded guesses and the total guess count
y=data['score_count_fraction'] # our dependent is the fraction of guesses that had the score pattern

r = linear_model.Ridge(alpha=1)
r.fit(X,y)

```

#### Results!

Now we can just look at the coefficients of the variables to see which ones most strongly predict a popular pattern. The top few guesses contain some [expected and unexpected](https://yougov.co.uk/topics/entertainment/articles-reports/2022/02/03/wordle-starter-words-hard-mode-and-x6-how-are-brit) entries. _(Note: that survey was only of British wordle players)_

```python
pd.DataFrame(list(zip(r.coef_, r.feature_names_in_)),
         columns=['coef', 'variable']).sort_values('coef',ascending=False).head(20)
```

### CRANE

On February 6, a popular video went somewhat viral positing that [CRANE was the best opener](https://www.youtube.com/watch?v=v68zYyaEmEA). Though this was later [recanted](https://www.youtube.com/watch?v=fRed0Xmc2Wg), it garned plenty of [media](https://kotaku.com/wordle-starting-word-math-science-bot-algorithm-crane-p-1848496404) [attention](https://www.forbes.com/sites/paultassi/2022/02/08/the-best-wordle-starter-word-a-first-guess-through-science/?sh=94c150a27909). CRANE was #5 on my list, how does it look before and after the video?

```python
data_recent = one_hot_encoded_data.query('wordle_num > 232')
X= data_recent.drop(
        columns=[ 'score_count_fraction', 'wordle_num']) # fit to the one hot encouded guesses and the total guess count
y=data_recent['score_count_fraction'] # our dependent is the fraction of guesses that had the score pattern

r = linear_model.Ridge(alpha=1)
r.fit(X,y)

```

```python
pd.DataFrame(list(zip(r.coef_, r.feature_names_in_)),
         columns=['coef', 'variable']).sort_values('coef',ascending=False).head(5)
```

Hello there, CRANE jumps to the second spot! What about before wordle 233?

```python
data_past = one_hot_encoded_data.query('wordle_num <= 232')
X= data_past.drop(
        columns=[ 'score_count_fraction', 'wordle_num']) # fit to the one hot encouded guesses and the total guess count
y=data_past['score_count_fraction'] # our dependent is the fraction of guesses that had the score pattern

r = linear_model.Ridge(alpha=1)
r.fit(X,y)




out = pd.DataFrame(list(zip(r.coef_, r.feature_names_in_)),
         columns=['coef', 'variable']).sort_values('coef',ascending=False)
out['guess_rank'] = out['coef'].rank(ascending=False)
out.query('variable == "crane"')
```

Prior to Wordle 233, CRANE ranked 808! Quite the turnaround. You can see this even in the cruder analysis.
